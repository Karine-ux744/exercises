```python
#VISUALIZAR A BASE DE DADOS
#importar base de dados
#importar a biblioteca responsável - "import (biblioteca) as (apelido)" é uma forma de dar apelido ao nome da biblioteca e assim posso usar esse apelido daqui em diante
#importar a base de dados: atribuir o arquivo a uma variável - variável = biblioteca.ação da biblioteca_formato do aquivo em que a biblioteca vai executar a ação("nome do arquivo")

import pandas as pd 


tabela = pd.read_csv("cancelamentos_sample.csv") 
display(tabela)
```

```python
#ARRUMAR A BASE DE DADOS/RESOLVE OS PROBLEMAS:

#informações inúteis: que não me ajudam a resolver o problema, não tem ligação com a solução do problema, não vou usar na minha análise. Informação que não ajuda, atrapalha
#informações no formato errado: informação que é número tem que estar no formato "int" ou "float" e textos no formato "object".
#informações vazias: informações vazias devem ser preenchidas ou eliminadas 

#informações inúteis - retirar
# a tabela que eu quero é a tabela inicial sem a informação que eu não quero. Para retirar essa informação (linha ou coluna) usar o comando "drop" - variável = variável.drop(o que eu quero remover = "nome do que eu quero remover")
#informações no formato errado - o comando "info" mostra informações sobre os dados da tabela
#informações vazias - para retirar as informações vazias: variavel = variavel.dropna()

tabela = tabela.drop(columns="CustomerID")
display(tabela)

tabela = tabela.dropna()
display(tabela.info())
```
